### RElATION - JOINPLOT - PAIRPLOT - HEATMAP

```
import seaborn as sns
import matplotlib.pyplot as plt
```

```
car = sns.load_dataset("car_crashes")
car.head(2)
```

```
sns.histplot(data=car,x='total', y='alcohol')
plt.show()
```

```
car.corr()
```

```
sns.heatmap(car.corr())
```

```
sns.heatmap(car.corr(), annot=True, cmap='coolwarm', cbar=True)
plt.show()
```

```
sns.heatmap(car.corr(), annot=True, cmap='coolwarm', cbar=True, linewidths=5, linecolor="black")
plt.show()
```
