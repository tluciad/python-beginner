**Muestreo sistemático**

```
def systematic_sampling(egresos, step):
  indexes = np.arange(0, len(egresos), step=step)
  systematic_sampling = egresos.iloc[indexes]
  return systematic_sampling
```

```
systematic_sample = systematic_sampling(egresos, 8)
systematic_sample 
```

**Muestreo estratificado en Python**

```
egresos['estratificado'] = egresos['desc_sector'] 
(egresos['estratificado'].value_counts()/len(egresos)).sort_values(ascending=False)
```

```
def data_estratificada(egresos, nombres_columnas_estrat, valores_estrat, prop_estrat, random_state = None):
  df_strat = pd.DataFrame(columns = egresos.columns)

  pos=-1
  for i in range(len(valores_estrat)):
    pos+=1
    if pos == len(valores_estrat) -1:
      ratio_len = len(egresos) - len(df_strat)
    else:
      ratio_len = int(len(egresos) * prop_estrat[i])
    df_filtrado = egresos[egresos[nombres_columnas_estrat] == valores_estrat[i]]
    df_temp = df_filtrado.sample(replace = True, n=ratio_len, random_state = random_state)
    df_strat = pd.concat([df_strat, df_temp])
    return df_strat
```

```
valores_estrat =['Gobierno', 'Transportes y Vialidad', 'Seguridad Ciudadana', 'Finanzas', 'Desarrollo Social', 'Educación', 'Salud', 'Medio Ambiente', 'Obras y Servicios', 'Desarrollo Económico', 'Protección Civil',
                 'Cultura', 'Jefatura de Gobierno', 'Contraloría General', 'Desarrollo Urbano y Vivienda', 'Consejería Jurídica', 'Trabajo y Fomento Al Empleo', 'Turismo', 'Mujeres', 'Desarrollo Rural y Equidad Para Las Comunidades',
                 'Sistema Público de Radiodifusión', 'Instituto de Planeación Democrática y Prospectiva', 'Órganos Autónomos', 'Poder Judicial', 'Deuda Pública', 'Subsidios', 'Fiscalía General de Justicia',
                 'Poder Legislativo', 'Procuraduría General de Justicia', 'Consejo de Evaluación de la Ciudad de México']
prop_estrat = [0.4, 0.07, 0.06, 0.06, 0.05, 0.04, 0.04, 0.03, 0.03, 0.02, 0.02, 0.02, 0.02, 0.02, 0.02, 0.015, 0.013, 0.011, 0.01, 0.005, 0.005, 0.0034, 0.0049, 0.0047, 0.0046, 0.0046, 0.0046, 0.0045, 0.00447, 0.00447]
df_strat =  data_estratificada(egresos, 'estratificado', valores_estrat, prop_estrat, random_state = 42)
df_strat
```
