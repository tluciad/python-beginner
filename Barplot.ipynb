```
import matplotlib.pyplot as plt
import numpy as np

Country = ['India', 'Japan', 'Mexico', 'Colombia', 'Germany']
Population = [3000, 800, 900, 1000, 300]

plt.bar(Country, Population)
plt.show
```

```
plt.bar(Country, Population, width=0.7, color=['blue', 'purple', 'gray', 'cyan', 'green'])
```

```
plt.bar(Country, Population, width=0.7, color=['#42687C', '#84A5B8', '#83DAF1', '#CBCBCB', '#707571'])
plt.xticks(rotation=45)
```

```
plt.bar(Country, Population, width=0.6, color=['#42687C', '#84A5B8', '#83DAF1', '#CBCBCB', '#707571'])
plt.xticks(rotation=45)
plt.title('Population by Countries')
plt.xlabel('Countries')
plt.ylabel('Population')
```
